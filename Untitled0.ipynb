```python
import pandas as pd

```

```python
dados = pd.read_csv('https://raw.githubusercontent.com/wandersonds/Expectativa-de-vida/refs/heads/main/dataset.csv?raw=true')
dados.head()
```

```python
dados.columns
```

```python
traducao_colunas = {
    'Country': 'País',
    'Year': 'Ano',
    'Status': 'Status_Desenvolvimento',
    'Life expectancy': 'Expectativa_Vida',
    'Adult Mortality': 'Mortalidade_Adultos',
    'infant deaths': 'Mortes_Infantis',
    'Alcohol': 'Consumo_Alcool',
    'percentage expenditure': 'Porcentagem_Gasto_Saude',
    'Hepatitis B': 'Hepatite_B',
    'Measles': 'Sarampo',
    'BMI': 'IMC',
    'under-five deaths': 'Mortes_Menores_5_Anos',
    'Polio': 'Poliomielite',
    'Total expenditure': 'Gasto_Total_Saude',
    'Diphtheria': 'Difteria',
    'HIV/AIDS': 'HIV_AIDS',
    'GDP': 'PIB',
    'Population': 'Populacao',
    'thinness  1-19 years': 'Magreza_1_19_Anos',
    'thinness 5-9 years': 'Magreza_5_9_Anos',
    'Income composition of resources': 'Indice_Recursos',
    'Schooling': 'Escolaridade'
}
dados.columns = dados.columns.map(traducao_colunas)
dados.head()
```

```python
from sklearn.model_selection import train_test_split

dados_treino, dados_val = train_test_split(dados, test_size=0.3, random_state=42, shuffle=True)

dados_treino.shape
```
